# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page=requests.get("https://www.imdb.com/list/ls091520106/")
soup=BeautifulSoup(page.content)


#Movie Name
name=[]
for i in soup.find_all('h3',class_="lister-item-header"):
    name.append(i.a.text)

#IDBM Rating    
rating=[]
for i in soup.find_all('div',class_="ipl-rating-star small"):
    rating.append(float(i.text))
    
#Year of Release
year=[]
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    x=i.text.replace("("," ")
    year.append(x.replace(")"," "))

# DataFrame

import pandas as pd
df= pd.DataFrame({"Name":name,"Rating":rating,"Year":year})
df
```

# 4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
soup=BeautifulSoup(page.content)


#banner--team-name
team=[]
for i in soup.find_all("td", class_="rankings-block__banner--team-name"):
    x=i.text.replace("\n","")
    team.append(x)
    
#banner--matches
matches=[]
for i in soup.find_all("td",class_="rankings-block__banner--matches"):
    matches.append(i.text)
       
#banner--points
points=[]
for i in soup.find_all("td", class_="rankings-block__banner--points"):
    points.append(i.text)
    
#banner--rating
Rating=[]
for i in soup.find_all("td", class_="rankings-block__banner--rating u-text-right"):
    Rating.append(i.text.replace("\n",""))

#table_team    
for i in soup.find_all("td", class_="table-body__cell rankings-table__team"):
    team.append(i.text.replace("\n",""))
    
#table Matche and points
matchepoint=[]
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matchepoint.append(i.text)
for i in range(0,len(matchepoint)-1,2):
    matches.append(matchepoint[i])
    points.append(matchepoint[i+1])
    
#table Ranking
for i in soup.find_all("td",class_="table-body__cell u-text-right rating"):
    Rating.append(i.text)
    
#DataFrame
import pandas as pd
df= pd.DataFrame({"Teams":team,"Matches":matches,"Point":points,"Rating":Rating})
df[:10]
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

# a) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating.

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
soup=BeautifulSoup(page.content)

#team Name
team=[]
for i in soup.find_all("span",class_="u-hide-phablet"):
    team.append(i.text)

#banner--matches
matches=[]
y=soup.find("td",class_="rankings-block__banner--matches")
matches.append(y.text)


#banner--points
points=[]
z=soup.find("td",class_="rankings-block__banner--points")
points.append(z.text)


# all matches and points value
matchespoints=[]
for i in soup.find_all("td", class_="table-body__cell u-center-text"):
    matchespoints.append(i.text)

for i in range(0,len(matchespoints)-1,2):
    matches.append(matchespoints[i])
    points.append(matchespoints[i+1])

#Allranking
ranking=[]
x=soup.find("td",class_="rankings-block__banner--rating u-text-right")
ranking.append(x.text.replace("\n",""))

for i in soup.find_all("td",class_="table-body__cell u-text-right rating"):
    ranking.append(i.text)



#dataFrame

df=pd.DataFrame({})
df['Team_name']=team[:10]
df['Matches']=matches[:10]
df['Points']=points[:10]
df['Ranking']=ranking[:10]
df 
```

# 6) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
page=requests.get("https://coreyms.com/")
soup=BeautifulSoup(page.content)

heading=[]
for i in soup.find_all("h2",class_="entry-title"):
    heading.append(i.text)
len(heading)

date=[]
for i in soup.find_all("time",class_="entry-time"):
    date.append(i.text)


content=[]
for i in soup.find_all("div",class_="entry-content"):
    content.append(i.p.text)

url=[]
for i in soup.find_all("iframe",class_="youtube-player"):
    url.append(i['src'])
len(url)

url.insert(4,"No data available")

#DataFrame

df=pd.DataFrame({})

df['Heading']=heading
df['Date']=date
df['Content']=content
df['URL']=url
df
```
